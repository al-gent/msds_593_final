```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('spotify.csv', encoding='iso-8859-1')
df_cleaned = df.dropna(subset=['Spotify Streams'])
len(df) - len(df_cleaned)
df_cleaned['Spotify Streams'] = df_cleaned['Spotify Streams'].str.replace(',','')
df_cleaned['Spotify Streams'] = df_cleaned['Spotify Streams'].astype('int')
df_cleaned.dtypes
df_cleaned.head()
```

```python
#tiktok views vs. spotify streams, youtube views, Pandora Streams, Soundcloud Streams, apple music playlist count, AirPlay Spins, SiriusXM Spins

platforms = ['Spotify Streams', 'YouTube Views', 'Pandora Streams', 'Soundcloud Streams', 
             'Apple Music Playlist Count', 'AirPlay Spins']
platforms1 = ['TikTok Views', 'Spotify Streams', 'YouTube Views', 'Pandora Streams', 'Soundcloud Streams', 
              'AirPlay Spins']
colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown', 'pink']

df_cleaned = df.dropna(subset=platforms1)

for column in platforms1:
    df_cleaned[column] = df_cleaned[column].astype('str')
    
for column in platforms1:
    df_cleaned[column] = df_cleaned[column].str.replace(',', '')
    df_cleaned[column] = pd.to_numeric(df_cleaned[column], errors='coerce')


for column in platforms1:
    df_cleaned[column] = df_cleaned[column].astype('int')
    
print(df_cleaned.dtypes)
df_cleaned.head()
```

```python
df_cleaned['YouTube Views'] = df_cleaned['YouTube Views'].astype('int')
fig, ax = plt.subplots(figsize=(6,4))

ax.scatter(df_cleaned['TikTok Views'], df_cleaned['YouTube Views'], label= 'YouTube Views' , color='brown', alpha=0.6)
ax.scatter(df_cleaned['TikTok Views'], df_cleaned['Spotify Streams'], label= 'YouTube Views' , color='pink', alpha=0.6)
ax.scatter(df_cleaned['TikTok Views'], df_cleaned['Soundcloud Streams'], label= 'YouTube Views' , color='blue', alpha=0.6)
ax.scatter(df_cleaned['TikTok Views'], df_cleaned['Pandora Streams'], label= 'YouTube Views' , color='green', alpha=0.6)
ax.scatter(df_cleaned['TikTok Views'], df_cleaned['Apple Music Playlist Count'], label= 'YouTube Views' , color='orange', alpha=0.6)
ax.scatter(df_cleaned['TikTok Views'], df_cleaned['AirPlay Spins'], label= 'YouTube Views' , color='yellow', alpha=0.6)
plt.show()
    
```

```python
correlations = df_cleaned.corr()
tiktok_correlations = correlations['TikTok Views'][platforms]

# Print the correlation coefficients
print(tiktok_correlations)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
tiktok_correlations.plot(kind='bar', color='skyblue', ax=ax)
ax.set_title('Correlation of TikTok Views with Other Platforms')
ax.set_ylabel('Correlation Coefficient')
ax.set_xlabel('Platform')
plt.xticks(rotation=45)
plt.show()
```

```python
# let's look into the artists in the top 10 tracks

df_cleaned.head(20)
top_10_artists = ['Tommy Richman', 'Kendrick Lamar', 'Artemas', 'Jack Harlow', 'Central Cee', 'Post Malone', 'Billie Eilish', 'Myke Towers', 'Taylor Swift', 'Sabrina Carpenter'] 

# create new df
top_10_df = df_cleaned[df_cleaned['Artist'].isin(top_10_artists)]

#how was their success distributed across various platforms this year?

#let's create a horizontal stacked bar chart showing percentages of streams on three major platorms 

```

```python
# Calculating Percentages

top_10_df['Total Streams'] = top_10_df['TikTok Views'] + top_10_df['Spotify Streams'] + top_10_df['YouTube Views']
top_10_df['TikTok Percentage'] = (top_10_df['TikTok Views'] / top_10_df['Total Streams']) * 100
top_10_df['Spotify Percentage'] = (top_10_df['Spotify Streams'] / top_10_df['Total Streams']) * 100
top_10_df['YouTube Percentage'] = (top_10_df['YouTube Views'] / top_10_df['Total Streams']) * 100
top_10_df = top_10_df.sort_values(by='TikTok Views')
```

```python
# Horizontal stacked bar graph 

fig, ax = plt.subplots(figsize=(10, 8))


ax.barh(top_10_df['Artist'], top_10_df['TikTok Percentage'], height=0.68, label="TikTok Percentage", color='#bababa')
ax.barh(top_10_df['Artist'], top_10_df['Spotify Percentage'], left=top_10_df['TikTok Percentage'], height=0.68, label="Spotify Percentage", color='#858585')
ax.barh(top_10_df['Artist'], top_10_df['YouTube Percentage'], left=top_10_df['TikTok Percentage'] + top_10_df['Spotify Percentage'], height=0.68, label="YouTube Percentage", color='#474747')

ax.set_xlabel('Percentage', size = 14)
ax.set_title("Percentage of Streams/Views Across Top 3 Platforms", pad = 20, size = 14)
ax.text(s="Top 10 Artists", color = "black" ,
         x=44, y=9.85, size = 13, alpha = 0.8)
ax.legend(loc = 'lower right', bbox_to_anchor=(1.2, 0.3)) 

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_facecolor(color='#f5efe1')
fig.patch.set_facecolor('#f5efe1')

plt.show()
```
